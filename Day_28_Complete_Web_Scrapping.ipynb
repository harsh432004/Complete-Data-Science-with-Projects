# Web Scraping

Library - BeautifulSoup

```python
url = "https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
# url = "https://harshfinancialservice.in" In this way you can add link of any website whose data you want to scrap
req = requests.get(url)
```

```python
req
```

```python
soup = BeautifulSoup(req.content)
```

```python
soup.prettify()
```

```python
arr = []
for i in soup.findAll('td'):
  print(i)
  arr.append(i)
```

```python
arr[2]
```

```python
re.sub('^<td>.*">|<td>|</td>|<.*>|\n',"",str(arr[2]))
```

```python
Film = []
Year =[]
Award = []
Nomination = []
count = 0

for i in soup.findAll('td'):
  i = re.sub('^<td>.*">|<td>|</td>|<.*>|\n',"",str(i))
  if count == 0:
    Film.append(i)
    count += 1
  elif count == 1:
    Year.append(i)
    count += 1
  elif count == 2:
    Award.append(i)
    count += 1
  else:
    count = 0
    Nomination.append(i)
```
